### 2. Какой части пассажиров удалось выжить?

### 6. Какое самое популярное женское имя на корабле?

```python
import re
def fname_extr(s):
    name_regex = re.compile(r'([^,]+), ([^\(]+)\(([^\)]+)\)')
    m_obj = re.match(name_regex, s)
    if m_obj:
        return m_obj.group(3).split()[0]
    else:
        return s.split(',')[1].strip().split()[1]
    
print(df[df['Sex']=='female']['Name'].apply(fname_extr).value_counts().index[0])
#fname_extr('Bonnell, Miss. Elizabeth')
```
